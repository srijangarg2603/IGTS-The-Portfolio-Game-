```python
import pandas as pd
data = {
    'Ticker': ['RELIANCE', 'TCS', 'INFY', 'HDFCBANK', 'WIPRO'],
    'Price': [2850.5, 3900.0, 1500.75, 1650.0, 420.3],
    'Volume': [1200000, 850000, 1100000, 950000, 700000],
    'Sector': ['Energy', 'IT', 'IT', 'Finance', 'IT']
}
df = pd.DataFrame(data)
print("Shape of DataFrame:", df.shape)
print("\nFirst 3 rows:")
print(df.head(3))
print("\nRows where Sector is 'IT':")
print(df[df['Sector'] == 'IT'])
df['Market_Cap_Proxy'] = df['Price'] * df['Volume']
print("\nDataFrame after adding Market_Cap_Proxy:")
print(df)
sorted_df = df.sort_values(by='Market_Cap_Proxy', ascending=False)
print("\nSorted DataFrame:")
print(sorted_df)
```

```python
mean_price = df['Price'].mean()
print("Mean Price:", mean_price)
it_volume = df[df['Sector'] == 'IT']['Volume'].sum()
print("Total IT Sector Volume:", it_volume)
highest_stock = df.loc[df['Market_Cap_Proxy'].idxmax(), 'Ticker']
print("Stock with Highest Market_Cap_Proxy:", highest_stock)
```

```python
import numpy as np
import matplotlib.pyplot as plt
daily_returns = np.random.normal(loc=0.0008, scale=0.015, size=252)
prices = 100 * np.cumprod(1 + daily_returns)
plt.plot(prices)
plt.title("Simulated Stock Price Series")
plt.xlabel("Trading Days")
plt.ylabel("Price")
plt.show()
print("Final Price:", prices[-1])
print("Peak Price:", np.max(prices))
running_max = np.maximum.accumulate(prices)
drawdowns = 100*(prices - running_max) / running_max
print("Maximum Drawdown:", np.min(drawdowns))
```

```python
import numpy as np
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
year1 = [42, 45, 50, 48, 55, 60, 58, 63, 67, 70, 75, 80]
year2 = [52, 49, 58, 61, 66, 71, 68, 74, 79, 83, 88, 95]
plt.subplot()
plt.plot(months, year1, marker='o', label='Year 1')
plt.plot(months, year2, marker='o', label='Year 2')
plt.title("Monthly Revenue Comparison")
plt.xlabel("Months")
plt.ylabel("Revenue (Lakhs)")
plt.legend()
plt.grid(True)
difference = np.array(year2) - np.array(year1)
colors = ['green' if d > 0 else 'red' for d in difference]
plt.subplot()
plt.bar(months, difference, color=colors)
plt.title("Difference in Revenue (Year2 - Year1)")
plt.xlabel("Months")
plt.ylabel("Difference")
plt.show()
```

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'Day': range(1, 11),
    'Price': [100, 103, np.nan, 108, 107, np.nan, 112, 115, np.nan, 120],
    'Volume': [500, 520, 510, np.nan, 530, 545, np.nan, 560, 570, 580]
})
print("Missing values in each column:")
print(df.isnull().sum())
df['Price'] = df['Price'].ffill()
df['Volume'] = df['Volume'].fillna(df['Volume'].mean())
print("\nDataFrame after filling missing values:")
print(df)
df['Return'] = ((df['Price'] - df['Price'].shift(1)) / df['Price'].shift(1)) * 100
print("\nDataFrame with Return column:")
print(df)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
returns = np.random.multivariate_normal(
    mean=[0.0005, 0.0008, 0.0003],
    cov=[[0.0004, 0.0001, 0.00005],
         [0.0001, 0.0006, 0.00008],
         [0.00005, 0.00008, 0.0003]],
    size=200
)
prices = 100 * np.cumprod(1 + returns, axis=0)
df = pd.DataFrame(prices, columns=['Asset_A', 'Asset_B', 'Asset_C'])
plt.plot(df['Asset_A'], label='Asset_A')
plt.plot(df['Asset_B'], label='Asset_B')
plt.plot(df['Asset_C'], label='Asset_C')
plt.title("Price Series of Three Assets")
plt.xlabel("Trading Days")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
returns_df = df.pct_change().dropna()
corr_matrix = returns_df.corr()
print("\nCorrelation Matrix:")
print(corr_matrix)
portfolio = df.mean(axis=1)
plt.plot(portfolio, label='Equally Weighted Portfolio', linewidth=2)
plt.legend()
plt.show()
```
