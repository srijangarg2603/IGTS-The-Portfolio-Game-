```python
#Question 2
def find_dominant_strategy(matrix):
    rows = len(matrix)
    cols = len(matrix[0])

    p1_dominant = None

    for r1 in range(rows):
        dominant = True

        for r2 in range(rows):
            if r1 == r2:
                continue

            for c in range(cols):
                if matrix[r1][c][0] <= matrix[r2][c][0]:
                    dominant = False
                    break

            if not dominant:
                break

        if dominant:
            p1_dominant = r1
            break

    p2_dominant = None

    for c1 in range(cols):
        dominant = True

        for c2 in range(cols):
            if c1 == c2:
                continue

            for r in range(rows):
                if matrix[r][c1][1] <= matrix[r][c2][1]:
                    dominant = False
                    break

            if not dominant:
                break

        if dominant:
            p2_dominant = c1
            break

    if p1_dominant is not None:
        print(f"Player 1 has a strictly dominant strategy: Row {p1_dominant}")
    else:
        print("Player 1 has no strictly dominant strategy.")

    if p2_dominant is not None:
        print(f"Player 2 has a strictly dominant strategy: Column {p2_dominant}")
    else:
        print("Player 2 has no strictly dominant strategy.")


# Example (i)
matrix1 = [
    [(6, 4), (8, 1)],
    [(4, 6), (7, 3)]
]

# Example (ii)
matrix2 = [
    [(-1, -1), (-3, 0)],
    [(0, -3), (-2, -2)]
]

# Example (iii)
matrix3 = [
    [(4, 4), (0, 2)],
    [(2, 0), (2, 2)]
]

find_dominant_strategy(matrix1)
find_dominant_strategy(matrix2)
find_dominant_strategy(matrix3)
```

```python
#Question 3
def iesds(matrix):
    active_rows = list(range(len(matrix)))
    active_cols = list(range(len(matrix[0])))

    changed = True
    iteration = 1

    while changed:
        changed = False

        print(f"\n--- Iteration {iteration} ---")
        row_removed = False

        for r1 in active_rows:
            for r2 in active_rows:

                if r1 == r2:
                    continue
                dominated = True

                for c in active_cols:
                    if matrix[r1][c][0] >= matrix[r2][c][0]:
                        dominated = False
                        break

                if dominated:
                    print(
                        f"Player 1: Row {r1} removed "
                        f"(strictly dominated by Row {r2})"
                    )

                    active_rows.remove(r1)
                    changed = True
                    row_removed = True
                    break

            if row_removed:
                break
        col_removed = False

        for c1 in active_cols:
            for c2 in active_cols:

                if c1 == c2:
                    continue
                dominated = True

                for r in active_rows:
                    if matrix[r][c1][1] >= matrix[r][c2][1]:
                        dominated = False
                        break

                if dominated:
                    print(
                        f"Player 2: Column {c1} removed "
                        f"(strictly dominated by Column {c2})"
                    )

                    active_cols.remove(c1)
                    changed = True
                    col_removed = True
                    break

            if col_removed:
                break

        if not changed:
            print("No more dominated strategies found.")

        iteration += 1
    print("\n=== Surviving Strategies ===")
    print("Remaining Rows for Player 1:", active_rows)
    print("Remaining Columns for Player 2:", active_cols)
    if len(active_rows) == 1 and len(active_cols) == 1:
        r = active_rows[0]
        c = active_cols[0]

        print("\nPredicted Outcome:")
        print(f"Cell ({r}, {c}) -> {matrix[r][c]}")


game = [
    [(1, 1), (-1, 2), (5, 0), (1, 1)],
    [(2, 3), (1, 2), (3, 0), (5, 1)],
    [(1, 1), (0, 5), (1, 7), (0, 1)]
]

iesds(game)
```

```python
#Question 4
def find_pure_nash_equilibria(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    p1_best_responses = set()
    p2_best_responses = set()
    print("P1 Best Responses:")

    for c in range(cols):

        max_payoff = max(matrix[r][c][0] for r in range(rows))

        best_rows = []

        for r in range(rows):
            if matrix[r][c][0] == max_payoff:
                p1_best_responses.add((r, c))
                best_rows.append(r)

        print(f"Against Column {c}: Row(s) {best_rows}")

    print("\nP2 Best Responses:")

    for r in range(rows):

        max_payoff = max(matrix[r][c][1] for c in range(cols))

        best_cols = []

        for c in range(cols):
            if matrix[r][c][1] == max_payoff:
                p2_best_responses.add((r, c))
                best_cols.append(c)

        print(f"Against Row {r}: Column(s) {best_cols}")

    nash_equilibria = []

    for r in range(rows):
        for c in range(cols):

            if ((r, c) in p1_best_responses and
                    (r, c) in p2_best_responses):

                nash_equilibria.append((r, c))

    print("\n=== Pure-Strategy Nash Equilibria ===")

    if nash_equilibria:
        for (r, c) in nash_equilibria:
            print(f"Cell ({r}, {c}) -> {matrix[r][c]}")
    else:
        print("No pure-strategy Nash Equilibrium found.")

#Example(i)
game1 = [
    [(-1, -1), (-3, 0)],
    [(0, -3), (-2, -2)]
]

#Example(ii)
game2 = [
    [(4, 4), (0, 2)],
    [(2, 0), (2, 2)]
]

#Example(iii)
game3 = [
    [(0, 0), (-1, 1)],
    [(1, -1), (-10, -10)]
]

#Example(iv)
game4 = [
    [(2, 1), (0, 0)],
    [(0, 0), (1, 2)]
]

#Example(v)
game5 = [
    [(4, 4), (0, 0), (1, 3)],
    [(0, 0), (5, 5), (0, 0)],
    [(3, 1), (0, 0), (6, 6)]
]

find_pure_nash_equilibria(game1)
find_pure_nash_equilibria(game2)
find_pure_nash_equilibria(game3)
find_pure_nash_equilibria(game4)
find_pure_nash_equilibria(game5)
```

```python
#Question 5
from itertools import product


def find_nash_equilibria(strategies, payoffs):

    num_players = len(strategies)


    all_profiles = list(product(*strategies))

    nash_equilibria = []

    print("Checking all strategy profiles...\n")

    for profile in all_profiles:

        current_payoffs = payoffs[profile]

        is_nash = True


        for player in range(num_players):

            current_strategy = profile[player]
            current_payoff = current_payoffs[player]


            for alt_strategy in strategies[player]:

                if alt_strategy == current_strategy:
                    continue


                deviated_profile = list(profile)
                deviated_profile[player] = alt_strategy
                deviated_profile = tuple(deviated_profile)

                deviated_payoff = payoffs[deviated_profile][player]


                if deviated_payoff > current_payoff:
                    is_nash = False

                    print(
                        f"{profile} is NOT Nash:"
                        f" Player {player + 1} can switch "
                        f"from {current_strategy} to {alt_strategy}"
                        f" ({current_payoff} -> {deviated_payoff})"
                    )

                    break

            if not is_nash:
                break


        if is_nash:
            nash_equilibria.append(profile)


    print("\n=== Pure Strategy Nash Equilibria ===")

    if nash_equilibria:
        for eq in nash_equilibria:
            print(f"{eq} -> Payoffs {payoffs[eq]}")
    else:
        print("No pure-strategy Nash Equilibrium exists.")


strategies = [
    ['C', 'D'],   # Player 1
    ['C', 'D'],   # Player 2
    ['C', 'D']    # Player 3
]

payoffs = {
    ('C', 'C', 'C'): [4, 4, 4],
    ('C', 'C', 'D'): [1, 1, 6],
    ('C', 'D', 'C'): [1, 6, 1],
    ('C', 'D', 'D'): [0, 3, 3],

    ('D', 'C', 'C'): [6, 1, 1],
    ('D', 'C', 'D'): [3, 0, 3],
    ('D', 'D', 'C'): [3, 3, 0],
    ('D', 'D', 'D'): [2, 2, 2]
}

find_nash_equilibria(strategies, payoffs)
```

This result is not very surprising because each player individually benefits from defecting when others cooperate. Even though mutual cooperation gives everyone a higher collective payoff (4,4,4), each player has an incentive to deviate for personal gain, leading the group toward universal defection (2,2,2). This illustrates the classic tension between individual incentives and collective welfare like in the prisoner's dilemma game.
